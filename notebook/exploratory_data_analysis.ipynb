### **Basic Business Health**

#### *Q1: How much revenue are we generating over time?*

```python
# --------------------------
# Calculating the Net Revenue
# --------------------------
net_revenue_df = df_analysis[(df_analysis['Cancelled'] == 0) & (df_analysis['Returned'] == 0)]
monthly_net_revenue = df_analysis.groupby('YearMonth')['TotalPrice'].sum().reset_index()

# Calculate monthly revenue growth
monthly_net_revenue['Revenue_Growth_Rate'] = monthly_net_revenue['TotalPrice'].pct_change()

# Calculate avg revenue growth
average_revenue_growth = monthly_net_revenue['Revenue_Growth_Rate'].dropna().mean()
print(f"Average Monthly Revenue Growth: {average_revenue_growth:.2%}")

# ----------------------
# Displaying the Figure
# ----------------------
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the data
ax.plot(monthly_net_revenue['YearMonth'], monthly_net_revenue['TotalPrice'], 
        color='royalblue', marker='o', linewidth=2.5, markersize=8)

# Format x-axis for dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)

# Add labels and title
ax.set_title('Monthly Net Revenue', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Month', fontsize=12, labelpad=10)
ax.set_ylabel('Revenue ()', fontsize=12, labelpad=10)

# Format y-axis with sterling sign and commas
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Add grid
ax.grid(True, alpha=0.3, linestyle='--')

# Add value annotations
for x, y in zip(monthly_net_revenue['YearMonth'], monthly_net_revenue['TotalPrice']):
    ax.annotate(f'{y:,.0f}', (x, y), textcoords="offset points", 
                xytext=(0,10), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

#### *Q3: Which countries generate the most revenue?*

#### *Q4: What are our best-selling products by revenue?*

```python
# -------------------------------------
# Calculating the best-selling product
# -------------------------------------
product_revenue = df_analysis.groupby(['StockCode', 'Description'])['TotalPrice'].sum().sort_values(ascending=False).reset_index()

# ----------------------
# Displaying the Figure
# ----------------------
top_products = product_revenue.head(10)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_products['Description'], top_products['TotalPrice'], 
                color='royalblue', edgecolor='black')

# Reverse y-axis to show highest revenue at top
plt.gca().invert_yaxis()

# Add value labels on bars with smart positioning
max_value = top_products['TotalPrice'].max()
for i, (bar, value) in enumerate(zip(bars, top_products['TotalPrice'])):
    formatted_value = f'{value:,.0f}'
    
    # If bar is wide enough, put text inside (right-aligned)
    if value > max_value * 0.1:  # If bar is more than 10% of max width
        x_position = value - (max_value * 0.01)  # Slightly inside the bar
        ha = 'right'
        color = 'white'
    else:  # If bar is too narrow, put text outside (left-aligned)
        x_position = value + (max_value * 0.01)  # Slightly outside the bar
        ha = 'left'
        color = 'black'
    
    plt.text(x_position, i, formatted_value, 
             va='center', ha=ha, color=color, fontweight='bold')

# Add titles and labels
plt.title('Top 10 Best-Selling Products by Revenue', fontsize=14, fontweight='bold')
plt.xlabel('Total Revenue ()')
plt.ylabel('Product')

plt.tight_layout()
plt.show()
```

#### *Q5: What is the average order value (AOV)?*

```python
# -------------------------------------
# Calculating the Average Order Value (AOV)
# -------------------------------------
invoice_revenue = df_analysis.groupby('InvoiceNo')['InvoiceValue'].first().reset_index()
aov = invoice_revenue['InvoiceValue'].mean()
print(f'The Average Order Value (AOV) are: {round(aov, 4)}')

# Calculate AOV overtime 
invoice_dates = df_analysis.groupby('InvoiceNo')['YearMonth'].min().reset_index()
invoice_revenue = invoice_revenue.merge(invoice_dates, on='InvoiceNo')
monthly_aov = invoice_revenue.groupby('YearMonth')['InvoiceValue'].mean().reset_index()

# ----------------------
# Displaying the Figure
# ----------------------
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the data - AOV over time
ax.plot(monthly_aov['YearMonth'], monthly_aov['InvoiceValue'], 
        color='royalblue', marker='s', linewidth=2.5, markersize=8, label='Monthly AOV')

# Format x-axis for dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)

# Add labels and title
ax.set_title('Monthly Average Order Value (AOV) Trend', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Month', fontsize=12, labelpad=10)
ax.set_ylabel('Average Order Value ()', fontsize=12, labelpad=10)

# Format y-axis with sterling sign and commas
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.2f}'))

# Add grid
ax.grid(True, alpha=0.3, linestyle='--')

# Add value annotations (formatted to 2 decimal places for currency)
for x, y in zip(monthly_aov['YearMonth'], monthly_aov['InvoiceValue']):
    ax.annotate(f'{y:,.2f}', (x, y), textcoords="offset points", 
                xytext=(0,10), ha='center', fontsize=9, fontweight='bold')

# Add a horizontal line for the overall AOV (calculated earlier)
ax.axhline(y=aov, color='red', linestyle='--', alpha=1.0, 
           label=f'Overall AOV: {aov:,.2f}')

# Add legend
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### **Returns, Cancellations & Risk**

#### *Q1: How frequent are cancellations and returns?*

```python
# -----------------------------
# Calculating the cancellation
# -----------------------------

# Getting 1 row per invoice 
invoice_level_cancellation = df_analysis.drop_duplicates('InvoiceNo')[['InvoiceNo', 'Cancelled']]

# Calculating the cancellation rate (%)
cancellation_rate = invoice_level_cancellation['Cancelled'].value_counts(normalize=True).rename('Percentage').reset_index()
cancellation_rate['Percentage'] *= 100

# Calculating the return rate (%)
return_rate = df_analysis['Returned'].value_counts(normalize=True).rename('Percentage').reset_index()
return_rate['Percentage'] *= 100


# ----------------------
# Displaying the Figure
# ----------------------
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Create labels for both charts
cancellation_labels = ['Not Cancelled', 'Cancelled']
return_labels = ['Not Returned', 'Returned']

# Cancellation Rate Pie Chart
axes[0].pie(cancellation_rate['Percentage'], 
            labels=cancellation_labels,
            autopct='%1.1f%%',
            colors=['royalblue', 'red'],
            startangle=90,
            textprops={'fontsize': 12, 'weight': 'bold'},
            explode=(0.05, 0.1))
axes[0].set_title('Invoice Cancellation Rate', fontsize=16, fontweight='bold', pad=20)

# Return Rate Pie Chart
axes[1].pie(return_rate['Percentage'], 
            labels=return_labels,
            autopct='%1.1f%%',
            colors=['green', 'yellow'],
            startangle=90,
            textprops={'fontsize': 12, 'weight': 'bold'},
            explode=(0.05, 0.1))
axes[1].set_title('Product Return Rate', fontsize=16, fontweight='bold', pad=20)

# Add overall title
fig.suptitle('Transaction Analysis: Cancellation vs Return Rates', fontsize=20, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()
```

#### *Q3: Do returns materially impact revenue?*

```python
# ----------------------------------------------------
# Calculating the revenue loss from returned products
# ----------------------------------------------------

# Calculating gross revenue
gross_revenue = df_analysis['TotalPrice'].sum()

# Calculating the revenue lost from returned products
returned_loss = df_analysis.loc[df_analysis['Returned'] == 1, 'TotalPrice'].sum()

# Calculating net revenue from return
net_rev_returned = gross_revenue - abs(returned_loss) # use absolute value since returned product has negative values (-)
revenue_impact = pd.DataFrame({
    'Metric': ['Gross Revenue', 'Revenue Lost to Returns', 'Net Revenue'],
    'Amount ()': [gross_revenue, abs(returned_loss), net_rev_returned]
})


# ----------------------
# Displaying the Figure
# ----------------------
plt.figure(figsize=(9, 6))

# Create bar chart with distinct colors
colors = ['#4CAF50', '#F44336', '#2196F3']  # Green, Red, Blue
bars = plt.bar(revenue_impact['Metric'], revenue_impact['Amount ()'], 
               color=colors, width=0.6)

# Add value labels on top of bars
for bar, amount in zip(bars, revenue_impact['Amount ()']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + (height * 0.01),
             f'{amount:,.0f}',  # Format with commas
             ha='center', va='bottom', fontsize=10)

# Format y-axis with  and commas
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x:,.0f}'))

# Add title and labels
plt.title('Revenue Impact of Returns', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Amount ()', fontsize=12, labelpad=10)

# Add grid for easier reading
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Calculate and show return percentage
return_pct = (abs(returned_loss) / gross_revenue) * 100
plt.text(0.5, 0.95, f'Returns: {return_pct:.1f}% of gross revenue', 
         transform=plt.gca().transAxes, ha='center', 
         fontsize=11, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```

### **RFM Segmentation**

#### *Q1: Can we segment customers using RFM?*

```python
# -----------------------------
# Creating RFM table
# -----------------------------

# Filter out negative revenue rows
df_positive_sales = df_analysis[df_analysis['TotalPrice'] > 0]

# Setting reference date 
reference_date = df_analysis['InvoiceDate'].max()

# Aggregating to customer level
rfm = df_positive_sales.groupby('CustomerID').agg(
    Recency = ('InvoiceDate', lambda x: (reference_date - x.max()).days),
    Frequency = ('InvoiceNo', 'nunique'),
    Monetary = ('TotalPrice', 'sum')
).reset_index()

# Creating RFM scores
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
rfm['RFM_Score'] = rfm['R_Score'].astype(int) + rfm['F_Score'].astype(int) + rfm['M_Score'].astype(int)


# --------------------------------------------------
# Segmenting the customers based on the RFM Score
# --------------------------------------------------

# Create the RFM segments
def create_rfm_segment(score):
    if score >= 12:
        return "High Value Customer"
    elif score >= 9:
        return "Loyal Customers"
    elif score >= 6:
        return "Potential Loyal Customer"
    elif score >= 4:
        return "At Risk"
    else:
        return "Lost Customers"

# Apply segmentation
rfm['RFM_Segment'] = rfm['RFM_Score'].apply(create_rfm_segment)

# Create a simpler high-value flag for analysis
rfm['High_Value'] = rfm['RFM_Score'] >= 9  # Top 40% based on typical distribution

# Save the RFM table as a csv
rfm.to_csv('../data/rfm_data.csv', index=False)

print("\nRFM Segment Distribution:")
print(rfm['RFM_Segment'].value_counts())
print(f"\nHigh Value Customers: {rfm['High_Value'].sum()} ({rfm['High_Value'].sum()/len(rfm)*100:.1f}%)")
```

```python
# ----------------------
# Displaying the Figure
# ----------------------

# Get segment distribution
segment_counts = rfm['RFM_Segment'].value_counts()
total_customers = len(rfm)

# Define colors for each segment (from best to worst)
colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#6A994E']

# Create the bar chart
plt.figure(figsize=(11, 6))

# Create bars
bars = plt.bar(segment_counts.index, segment_counts.values, 
               color=colors, edgecolor='black', linewidth=1.2, alpha=0.8)

# Customize the chart
plt.title('RFM Customer Segments Distribution', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Customer Segment', fontsize=12, fontweight='bold')
plt.ylabel('Number of Customers', fontsize=12, fontweight='bold')
plt.xticks(rotation=0, fontsize=11)

# Add grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--')

# Add count and percentage labels on top of each bar
for bar in bars:
    height = bar.get_height()
    count = int(height)
    percentage = (height / total_customers) * 100
    
    # Position the label slightly above the bar
    plt.text(bar.get_x() + bar.get_width()/2., height + (total_customers * 0.01),
             f'{count}\n({percentage:.1f}%)', 
             ha='center', va='bottom', fontsize=10, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8, edgecolor='gray'))

# Add total customers annotation
plt.text(0.80, 0.95, f'Total Customers: {total_customers:,}', 
         transform=plt.gca().transAxes, fontsize=11,
         bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.9))

# Adjust layout
plt.tight_layout()
plt.show()
```

#### *Q2: Which customer segments generate most revenue?*

```python
# Calculate total monetary value by segment
segment_revenue = rfm.groupby('RFM_Segment')['Monetary'].sum().sort_values(ascending=False)
segment_counts = rfm.groupby('RFM_Segment').size()

print("Total Revenue by Segment:")
for segment, revenue in segment_revenue.items():
    count = segment_counts[segment]
    avg_revenue = revenue / count
    print(f"{segment}: {revenue:,.2f} ({count} customers, avg: {avg_revenue:,.2f})")

# Calculate percentage of total revenue
total_revenue = rfm['Monetary'].sum()
segment_percentage = (segment_revenue / total_revenue * 100).round(1)

print(f"\nTotal Revenue: ${total_revenue:,.2f}")
print("\nPercentage of Total Revenue by Segment:")
for segment, percentage in segment_percentage.items():
    print(f"{segment}: {percentage}%")
```
